```python
import json
import urllib.error
import urllib.parse
import urllib.request
from tciaclient import TCIAClient
####################################  Function to print server response #######


def printServerResponse(response):
    """
    Print and return server rersponse as byte string
    """
    if response.getcode() == 200:
        print("Server Returned:\n")
        resp = response.read()
        print(resp)
        print("\n")
    else:
        print("Error: " + str(response.getcode()))
        resp = None
    return resp
```

```python
tcia_client = TCIAClient(
    baseUrl="https://services.cancerimagingarchive.net/services/v4", resource="TCIA")

# find all datasets on site
try:
    response = tcia_client.get_collection_values(
        outputFormat="json")
    resp = printServerResponse(response)
except urllib.error.HTTPError as err:
    print("Errror executing program:\nError Code: ",
          str(err.code), "\nMessage:", err.read())
```

```python
# make list of collections names
resp = json.loads(resp)
# print(resp)
datasets_names = []
for r in resp:
    for v in r.values():
        datasets_names.append(v)
print(datasets_names)
```

```python
# takes patients information from "PROSTATE-MRI" collection
try:
    response = tcia_client.get_patient_study(
        collection="PROSTATE-MRI", patientId=None, studyInstanceUid=None, outputFormat="json")
    printServerResponse(response)

except urllib.error.HTTPError as err:
    print("Errror executing program:\nError Code: ",
          str(err.code), "\nMessage:", err.read())

```

```python
# takes series information from "PROSTATE-MRI" collection
try:
    response = tcia_client.get_series(collection="PROSTATE-MRI")
    printServerResponse(response)

except urllib.error.HTTPError as err:
    print("Errror executing program:\nError Code: ", str(err.code), "\nMessage:", err.read())
```
